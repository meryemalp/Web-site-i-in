```python
#Import the random library
import random as rd
#create a list containing the three actions of the game.
action_list = ['taş', 'kağıt', 'makas']
#Set the scores of players to 0
computer_score = 0
player_score = 0

#Ask the user how many rounds they want to play
total_rounds = input("Kaç round oynamak istiyorsun? Bir sayı girin: ")

#Add a round_counter that is 0 at the beginning
round_counter = 0

#Write a while loop and put the game inside
while True:

  #Increase round_counter by and print it
  round_counter += 1
  print("Round sayısı:", round_counter)


  #Select a random action for computer
  computer_choice = rd.choice(action_list)

  #Ask player to choose an action
  player_choice = input("Seçimini gir ( taş, kağıt ya da makas) : ")

  #Print the players choices
  print("Bilgisayar:", computer_choice)
  print("Oyuncu:", player_choice)


  #tie condition
  if computer_choice == player_choice:
      print("Tüh! Aynısını seçtiniz.")


  #Remaining conditions
  elif computer_choice == 'kağıt':
    if player_choice == 'taş':
        print("Kazanan: Bilgisayar")
        computer_score += 1
    else:
        print("Winner is: Player")
        player_score += 1 

  elif  computer_choice == 'taş':
    if player_choice == 'kağıt':
        print("Kazanan: Oyuncu")
        player_score += 1
    else:
        print("Kazanan: Bilgisayar")
        computer_score += 1

  elif computer_choice == 'makas':
    if player_choice == 'kağıt':
        print("Kazanan: Bilgisayar")
        computer_score += 1
    else:
        print("Kazanan: Oyuncu")
        player_score += 1

  if round_counter == int(total_rounds):
    break

if computer_score == player_score:
  print("Durum eşit!", computer_score, ':', player_score)
elif computer_score > player_score:
  print("Bilgisayar kazandı!!!", computer_score, ':', player_score)
elif computer_score < player_score:
  print("Oyuncu kazandı!!!", computer_score, ':', player_score)
  
```
